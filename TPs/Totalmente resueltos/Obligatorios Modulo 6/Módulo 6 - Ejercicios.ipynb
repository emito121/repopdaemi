- 5.4) A partir de un array 1D obtenga un array con los valores correspondientes al "ranking" que ocupan los valores del array original, sin cambiar el orden (Hint: Se recomienda usar *numpy.argsort()*). Vea el ejemplo.

    ```python

    ## a es array([ 6, 19, 14, 10,  7,  6, 18, 10, 10,  3])
    ```

````python
import numpy as np
def mediaMovil(vector, factor):
    '''
    Devuelve un numpy array filtrado a partir de una media móvil.
    Argumentos:
        -vector: numpy array a ser filtrado
        -factor: número de muestras a tomar para la promediación
    '''
    nuevo_vector = np.copy(vector)
    for i, k in enumerate(vector):
        try:
            nuevo_vector[i] = np.mean(vector[i-factor:i])
        except:
            pass
    return nuevo_vector
````
